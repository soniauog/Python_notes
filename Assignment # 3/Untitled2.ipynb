```python
#Question:7
data = input("Enter marks of your 5 subjects")
n = data.split()
sum = 0
for i in range(len(n)):
    sum = sum + int(n[i])
print("*************Mark Sheet*************")
print("\tTotal Marks = ",sum)
prcntage =(sum/500)*100
print("\tPercentage = ",prcntage)
if prcntage>=85 and prcntage<=100:
    print("\tGrade = A+")
elif prcntage>=75 and prcntage<85:
    print("\tGrade = B+")
elif prcntage>=65 and prcntage<75:
    print("\tGrade = c")
elif prcntage>=55 and prcntage<65:
    print("\tGrade = D")
elif prcntage>=45 and prcntage<55:
    print("\tGrade = E")
else:
    print("\tGrade = F \n\tTry Again")
```
